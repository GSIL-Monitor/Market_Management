```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException

chrome_options = Options()
# chrome_options.add_argument('--headless')
chrome_options.add_argument('--disable-gpu')
chrome_options.binary_location = r'C:\Program Files (x86)\Google\Chrome\Application\chrome.exe'
browser = webdriver.Chrome(chrome_options=chrome_options)
browser.get('https://www.alibaba.com/')
```

```python
browser.quit()
```

```python
keyword = 'mink eyelash'
words = '_'.join(keyword.split(' '))
url_showroom = 'https://www.alibaba.com/showroom/'+words+'.html'
browser.get(url_showroom)
```

```python
items = browser.find_elements_by_css_selector('div.item-main')
print('产品数量: ', len(items))

title_list = []
keywords_list = []
counter = 1
for item in items:
    print('No. '+ str(counter) + '.')
    counter = counter + 1
    
    title = item.find_element_by_css_selector('div.item-info h2>a').get_attribute('title')
    title_list.append(title)
    print('Title: ', title)
    
    try:
        tag = item.find_element_by_css_selector('div.tags')
    except NoSuchElementException:
        print()
        continue
    
    
    tag = tag.get_attribute('innerHTML').replace('Tags:', '')
    keywords = []
    for word in tag.split('|')[0:-1]:
        keywords.append(word.strip())
    keywords_list.append(keywords)
    print('Keywords: ', ', '.join(keywords))
    print()
```
